```python
new_df=df[['Song-Name','tags']]
new_df
```

```python
#A lambda function is an anonymous function. Here, it takes x (an individual element from the tags column) and converts it to lowercase using the .lower()
new_df['tags']=new_df['tags'].apply(lambda x:x.lower()) 
new_df
```

```python
#CountVectorizer is a tool that converts a collection of text documents into a matrix of 
# token counts,
#meaning it counts the number of occurrences of each word in the text data.
#max_features=2000 specifies that only the 2,000 most frequent words (tokens) should be included in the vocabulary.
#Less frequent words will be ignored. 
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2000) 
```

```python
 #fit:  it identifies the unique words in the dataset.
#transform: Transforms the text data into a numerical representation, often a sparse matrix.
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
cv.get_feature_names_out()
```

Cosine similarity is a metric used to measure how similar two vectors are by computing the cosine of the angle between them.
A value of 1 means the vectors are identical, 
while a value of 0 means they are orthogonal (completely dissimilar).

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])
```

```python
new_df.rename(columns={'Song-Name':'title'},inplace=True)
```

```python
def recommend(music):
    ##checks which row in the new_df DataFrame has the song title equal 
    ## to the music argument.
    ## .index[0] retrieves the index of that row.
    ## This music_index is the row number in the DataFrame corresponding to the input song.
    music_index=new_df[new_df['title']==music].index[0]

    ### similarity: This is the similarity matrix created earlier using cosine similarity.
    ### Each entry in the matrix represents the cosine similarity score between two songs.
    ### similarity[music_index]: Retrieves the row of the similarity matrix that 
    # corresponds to the input song (music).
    distances=similarity[music_index]

    ### list(enumerate(distances)): This creates a list of tuples where each tuple 
    # contains the index of a song and its corresponding similarity score 
    # to the input song. The format would look like 
    # [(0, score1), (1, score2), (2, score3), ...].

    ### sorted(..., reverse=True, key=lambda x: x[1]): This sorts the list of tuples 
    # in descending order based on the similarity score (x[1] is the score). 
    # The most similar songs will appear first.

    ### [1:6]: The first song (index 0) is the input song itself 
    # (which has a similarity score of 1 with itself), 
    # so [1:6] selects the next 5 most similar songs.
    music_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in music_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Proper Patola')
```
